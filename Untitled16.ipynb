```python
# Lowercase all labels
df['label'] = df['label'].astype(str).str.strip().str.lower()

# Map variations to standard form
mapping = {
    "pos": "positive", "positive": "positive",
    "neg": "negative", "negative": "negative",
    "neu": "neutral", "neutral": "neutral"
}
df['label'] = df['label'].map(mapping).fillna(df['label'])

print("Unique labels after normalization:", df['label'].unique())
print("Counts:\n", df['label'].value_counts())
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download missing resources
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
nltk.download("wordnet")

STOPWORDS = set(stopwords.words("english"))
LEMMA = WordNetLemmatizer()

def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", " ", text)      # remove links
    text = re.sub(r"\S+@\S+", " ", text)             # remove emails
    text = re.sub(r"[^a-z0-9\s]", " ", text)         # keep only alphanum
    tokens = word_tokenize(text)
    tokens = [LEMMA.lemmatize(t) for t in tokens if t not in STOPWORDS]
    return " ".join(tokens)

# Apply cleaning on reply column
df['clean_text'] = df['reply'].astype(str).apply(clean_text)

# Preview cleaned dataset
df[['reply','clean_text','label']].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Encode labels
le = LabelEncoder()
y_enc = le.fit_transform(df['label'])

# Features and labels
X = df['clean_text']
y = y_enc

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score, classification_report

# TF-IDF + Logistic Regression pipeline
tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1,2))
clf = LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42)

pipe = Pipeline([
    ("tfidf", tfidf),
    ("clf", clf)
])

# Train
pipe.fit(X_train, y_train)

# Predict
y_pred = pipe.predict(X_test)

# Evaluate
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred))
print("Logistic Regression F1:", f1_score(y_test, y_pred, average="macro"))
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
import lightgbm as lgb

# Transform text to TF-IDF for LGB
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf  = tfidf.transform(X_test)

lgb_clf = lgb.LGBMClassifier(n_estimators=200, random_state=42)
lgb_clf.fit(X_train_tfidf, y_train)

# Predict
y_pred_lgb = lgb_clf.predict(X_test_tfidf)

# Evaluate
print("LightGBM Accuracy:", accuracy_score(y_test, y_pred_lgb))
print("LightGBM F1:", f1_score(y_test, y_pred_lgb, average="macro"))
print(classification_report(y_test, y_pred_lgb, target_names=le.classes_))
```

```python
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
import torch
import numpy as np
from sklearn.metrics import accuracy_score, f1_score

# Use raw 'reply' text for transformer
df_train, df_eval = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=42)

# Encode labels
df_train['label_enc'] = le.transform(df_train['label'])
df_eval['label_enc'] = le.transform(df_eval['label'])

# HuggingFace datasets
ds_train = Dataset.from_pandas(df_train[['reply','label_enc']].rename(columns={'reply':'text','label_enc':'label'}))
ds_eval = Dataset.from_pandas(df_eval[['reply','label_enc']].rename(columns={'reply':'text','label_enc':'label'}))

# Tokenizer
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_fn(batch):
    return tokenizer(batch['text'], truncation=True, padding="max_length", max_length=128)

ds_train = ds_train.map(tokenize_fn, batched=True)
ds_eval = ds_eval.map(tokenize_fn, batched=True)

# Format for PyTorch
ds_train.set_format(type="torch", columns=["input_ids","attention_mask","label"])
ds_eval.set_format(type="torch", columns=["input_ids","attention_mask","label"])
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load tokenizer and model (feature-based, no fine-tuning)
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
bert_model = AutoModel.from_pretrained(model_name).to(device)
bert_model.eval()

# Helper function to get embeddings
def get_embeddings(text_list):
    inputs = tokenizer(
        text_list,
        padding=True,
        truncation=True,
        max_length=128,
        return_tensors="pt"
    ).to(device)
    with torch.no_grad():
        outputs = bert_model(**inputs)
        embeddings = outputs.last_hidden_state[:,0,:]  # CLS token
    return embeddings.cpu().numpy()

# Get embeddings
X_train_emb = get_embeddings(df_train['clean_text'].tolist())
X_test_emb = get_embeddings(df_eval['clean_text'].tolist())

# Train a classifier (fast)
clf = LogisticRegression(max_iter=1000, class_weight="balanced")
clf.fit(X_train_emb, df_train['label_enc'])

# Predict
y_pred = clf.predict(X_test_emb)

# Metrics
print("Accuracy:", accuracy_score(df_eval['label_enc'], y_pred))
print("Macro F1:", f1_score(df_eval['label_enc'], y_pred, average="macro"))
print(classification_report(df_eval['label_enc'], y_pred, target_names=le.classes_))
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Predict on test embeddings
y_pred = clf.predict(X_test_emb)

# Accuracy
acc = accuracy_score(df_eval['label_enc'], y_pred)
macro_f1 = f1_score(df_eval['label_enc'], y_pred, average='macro')

print(f"Accuracy: {acc:.4f}")
print(f"Macro F1: {macro_f1:.4f}")

# Detailed classification report
print(classification_report(
    df_eval['label_enc'],
    y_pred,
    target_names=le.classes_
))
```

```python
import pickle
import torch
from transformers import AutoTokenizer, AutoModel
import numpy as np

# Load trained classifier
with open("clf_model.pkl", "rb") as f:
    clf = pickle.load(f)

with open("label_encoder.pkl", "rb") as f:
    le = pickle.load(f)

# DistilBERT model and tokenizer
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
bert_model = AutoModel.from_pretrained(model_name)
bert_model.eval()
device = "cuda" if torch.cuda.is_available() else "cpu"
bert_model.to(device)
```

```python
def predict_reply(text):
    # Tokenize and get embeddings
    inputs = tokenizer([text], padding=True, truncation=True, max_length=128, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = bert_model(**inputs)
        emb = outputs.last_hidden_state[:,0,:].cpu().numpy()  # CLS token

    # Predict with classifier
    pred_label_enc = clf.predict(emb)[0]
    pred_prob = np.max(clf.predict_proba(emb))
    pred_label = le.inverse_transform([pred_label_enc])[0]

    return {"label": pred_label, "confidence": round(float(pred_prob), 2)}
```

```python
test_text = "Could u clarify features included?."
prediction = predict_reply(test_text)
print(prediction)
```
